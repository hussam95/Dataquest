```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
exchange_rates= pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.info()
exchange_rates.rename(columns={'[US dollar ]':'US_dollar', 'Period\\Unit:':'Time'}, inplace=True)
exchange_rates
```

```python
exchange_rates['Time']=pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.info()
```

```python
exchange_rates.sort_values('Time', inplace=True)
```

```python
exchange_rates
```

```python
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
euro_to_dollar=exchange_rates[['Time','US_dollar']]
```

```python
euro_to_dollar
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar=euro_to_dollar[euro_to_dollar['US_dollar']!='-']
```

```python
euro_to_dollar
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar['US_dollar']=euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar.info()
```

```python
euro_to_dollar
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
```

```python
euro_to_dollar['rolling_mean']=euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
euro_to_dollar.info()
```

```python
plt.plot(euro_to_dollar['Time'],euro_to_dollar['rolling_mean'])
```

```python
euro_to_dollar['rolling_mean']
```

## Euro-USD Performance During Financial Crisis 2007-2008

```python
euro_to_dollar_fc = euro_to_dollar[
              (euro_to_dollar['Time'].dt.year==2006) |
              (euro_to_dollar['Time'].dt.year==2007) |
              (euro_to_dollar['Time'].dt.year==2008) |
              (euro_to_dollar['Time'].dt.year==2009) |
              (euro_to_dollar['Time'].dt.year==2010)
              
              ]
euro_to_dollar_focus=euro_to_dollar[
              (euro_to_dollar['Time'].dt.year==2007) |
              (euro_to_dollar['Time'].dt.year==2008) 
              
              
              
              ]
```

```python
euro_to_dollar_fc
```

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(euro_to_dollar_fc['Time'],euro_to_dollar_fc['rolling_mean'],  color='green', alpha=0.3)
ax.plot(euro_to_dollar_focus['Time'],euro_to_dollar_focus['rolling_mean'],  color='red', linewidth=3)

ax.set_yticks([1.2,1.3,1.4,1.5])
ax.tick_params(left=False,bottom=False)
ax.text(-0.05, 1, 
        'Euro-USD rate peaked at 1.59 during 2007-08 financial crisis',
        verticalalignment='center',
        transform=ax.transAxes, fontsize=12,weight='bold')
ax.text(-0.05, 0.95, 
        'Euro-USD exchange rates 2006-11',
        verticalalignment='center',
        transform=ax.transAxes, fontsize=12)
ax.axvspan(xmin=733112.0, xmax=733302.0, ymin=0.09, ymax=.98,
           alpha=0.3, color='grey')
ax.text(-0.05, -0.13, '©DATAQUEST' + ' '*112 + 'Source: European Central Bank' ,
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10,transform=ax.transAxes)
for location in ['top','left','right','bottom']:
    ax.spines[location].set_visible(False)
plt.grid(alpha=0.3,color='grey')
plt.show()
```
