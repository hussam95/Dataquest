# Reading and Cleaning Data Set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
star_wars= pd.read_csv("star_wars.csv", encoding= "ISO-8859-1")
pd.set_option('display.max_columns', None)
star_wars.head(5)
```

```python
cols=["Have you seen any of the 6 films in the Star Wars franchise?",
      "Do you consider yourself to be a fan of the Star Wars film franchise?"]
for c in cols:
    star_wars[c]=star_wars[c].map({"Yes":True, "No":False, np.nan:False, "Response":"Response"})
    
```

```python
star_wars[cols]=star_wars[cols]
star_wars.head(3)
```

```python
star_wars.columns.tolist()
```

```python
cols=['Which of the following Star Wars films have you seen? Please select all that apply.',
 'Unnamed: 4',
 'Unnamed: 5',
 'Unnamed: 6',
 'Unnamed: 7',
 'Unnamed: 8']

num=1
for c in cols:
    star_wars[c]=star_wars[c].map({'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,          
    'Star Wars: Episode VI Return of the Jedi': True, 
     np.NaN: False})
    for i in range(1):
        star_wars=star_wars.rename(columns={c:"seen_{}".format(num)})
        num+=1
        
    

star_wars
```

```python
star_wars[star_wars.columns[9:15]]=star_wars.iloc[1:][star_wars.columns[9:15]].astype(float)

cols=star_wars.columns[9:15]
num=1
for c in cols:
    for i in range(1):
        star_wars=star_wars.rename(columns={c:"ranking_{}".format(num)})
        num+=1
    
```

```python
#Eliminating row at index zero that does not correspond to any survey participant
star_wars=star_wars.iloc[1:][star_wars.columns[0:]]
```

# Rankings and #Views Per Movie/ All Respondents

```python
%matplotlib inline
star_wars[star_wars.columns[9:15]].mean().sort_values().plot.bar(color=['green','yellow','violet','pink','orange','red'],
                                                                 title="Avg. Ranking per Movie/All Respondents")
```

**Based on the bar graph, we find that out of the 6 movies in the franchise, the respondents liked these parts the most (in order of most liked to least liked: 1=highest ranked, 6= lowest ranked)**  
- Star Wars: Episode V The Empire Strikes Back
- Star Wars: Episode VI Return of the Jedi
- Star Wars: Episode IV A New Hope

**It is interesting to note the order of rankings that fans assigned to each movie in the Star Wars series. The recent releases got better rating than the earlier parts of the series. Studies have shown that audience likes the end and beginings of an event more than the middle parts. This finding is also applicable in the case of Star Wars series as people ranked the end and start better than they ranked the middle parts of the series.**

```python
star_wars[star_wars.columns[3:9]].sum().sort_values().plot.bar(color=['green','yellow','violet','pink','orange','red'],
                                                               title="Views per Movie/All Respondents")
```

**While plotting the number of times each movie of the Star Wars series is seen by the respondents, it is clear that 5th is seen the most followed by the 6th part. We already know that these movies got better audience ranking which could have led more people to watch them.** 

# Rankings and #Views Per Movie w.r.t Respondent's Gender 

```python
#Compartmentalizing data set w.r.t gender
male=star_wars[star_wars["Gender"]=="Male"]
female=star_wars[star_wars["Gender"]=="Female"]
# Calculating Mean Male Rankings for 6 Movies
male_rankings=male[male.columns[9:15]].mean().sort_values()
# Calculating Mean Male Rankings for 6 Movies
female_rankings=female[female.columns[9:15]].mean().sort_values()
# Converting Male Rankings Series to a Dataframe
male_rankings=male_rankings.to_frame(name="Male")
# Adding Female Rankings Column to male_rankings df
male_rankings["Female"]=female_rankings
# Creating new df both_rank to hold mean rankings for both female and male participants
both_rank=male_rankings
# Displaying both_rank df
both_rank
```

```python
both_rank.plot.bar(title="Avg. Ranking per Movie w.r.t Gender")
```

**Females gave less ratings (remember it's an inverted scale) to 4 out of 6 Star Wars Movies which shows they like the series a little less than their male counterparts**

```python
# Summing views for each movie for male respondents + sorting w.r.t descending order of #views + making a new df 'views' 
views=male[male.columns[3:9]].sum().sort_values(ascending=False).to_frame(name="Male")
# Adding a new col. to df 'views' called Female that contains total views per movie for females 
views["Female"]=female_views=female[female.columns[3:9]].sum().sort_values(ascending=False)
```

```python
views
```

```python
views.plot.bar(title="Views Per Movie w.r.t Gender")
```

**Gender-based segmentation shows that men from the survey watched each movie of Star Wars Series more than their female counterparts.**
